```python
features = ['altitude', 'distance', 'gender_', 'heart_rate', 'latitude', 'longitude', 'speed']
X = run_df.loc[:, features]
X.head()
```

```python
# uknn_model = NearestNeighbors(n_neighbors=10).fit(X)
```

```python
def recommend(user_id, n_recommendations=5):
    global features
    
    temp_df = run_df.loc[run_df.userId == user_id, features]
    temp_df = temp_df.mean()
    
    # Fit model
    X = run_df.loc[run_df.userId != user_id, features]
    uknn_model = NearestNeighbors(n_neighbors=n_recommendations).fit(X)
    
    # get n nearest neighbors
    distances, indices = uknn_model.kneighbors(temp_df.values.reshape(1, -1))
    
    # get result
    result_df = run_df.loc[indices[0]]
    
    return result_df
```

```python
result_df = recommend(478585, n_recommendations=5)
```
