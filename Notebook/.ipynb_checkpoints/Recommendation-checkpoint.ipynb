```python
run_df.shape
```

```python
features = ['altitude', 'distance', 'gender', 'heart_rate', 'latitude', 'longitude', 'speed']
X = run_df.loc[:, features]
X.head()
```

```python
scaled_run = MinMaxScaler().fit_transform(run_df.loc[:, features])
scaled_run_df = pd.DataFrame(scaled_run, columns=features)
scaled_run_df['id'] = run_df.id
scaled_run_df['userId'] = run_df.userId
scaled_run_df.head()

scaled_bike = MinMaxScaler().fit_transform(run_df.loc[:, features])
scaled_bike_df = pd.DataFrame(scaled_bike, columns=features)
scaled_bike_df['id'] = bike_df.id
scaled_bike_df['userId'] = bike_df.userId
```

```python
def recommend(user_id, n_recommendations=5):
    global features
    
    # Select user data and scale it to range (0, 1)
    temp_df = scaled_run_df.loc[run_df.userId == user_id, features].mean()
    
    # Fit model
    X = scaled_run_df.loc[run_df.userId != user_id, features]
    uknn_model = NearestNeighbors(n_neighbors=n_recommendations).fit(X)
    
    # get n nearest neighbors
    distances, indices = uknn_model.kneighbors(temp_df.values.reshape(1, -1))
    
    # get result
    result_df = run_df.loc[indices[0]]
    result_df['neighbor_dist'] = distances[0]
    result_df['for'] = [user_id for i in range(n_recommendations)]
    
    return result_df
```

```python
result_df = recommend(478585, n_recommendations=5)
```

## Recommendation by nearest route location with KMeans & Nearest Neighbors

```python
run_df.head()
```

```python
X = scaled_run_df.loc[:, ['longitude', 'latitude']]
km_model = KMeans(n_clusters=4, random_state=88)
km_model.fit(X)
predictions = km_model.predict(X)
run_df['cluster'] = predictions
scaled_run_df['cluster'] = predictions
```

```python
plt.scatter(X.longitude, X.latitude, c=predictions, cmap='rainbow')
plt.scatter(km_model.cluster_centers_[:, 0], km_model.cluster_centers_[:, 1], marker='X', c='black')
```

```python
def recommend_byclustloc(user_id, n_recommendations=5):
    global km_model
    global scaled_run_df
    
    # Get user's cluster
    temp_df = scaled_run_df.loc[run_df.userId == user_id, ['longitude', 'latitude']]
    temp_df = temp_df.mean().values.reshape(1, -1)
    predictions = km_model.predict(temp_df)
    print(predictions)
    
    # Get user friends
    mask = (scaled_run_df['cluster'] == predictions[0]) & (run_df.userId != user_id)
    cluster_member = scaled_run_df.loc[mask].reset_index(drop=True)
    
    temp_df2 = scaled_run_df.loc[mask].reset_index(drop=True)
    
    # Prepare friends
    X = cluster_member.loc[cluster_member.userId != user_id, ['longitude', 'latitude']]
    plt.scatter(X.longitude, X.latitude)
    plt.show()
    
    # Get the nearest friend from cluster
    model = NearestNeighbors(n_neighbors=n_recommendations).fit(X)
    predictions = model.kneighbors(temp_df)
    
    return predictions, temp_df2, temp_df, X
```

```python
(distances, indices), temp_df, user, X = recommend_byclustloc(478585, 10)
predictions, indices
```

```python
temp_df.loc[indices[0]]
```

```python
plt.scatter(X.longitude, X.latitude)
plt.scatter(temp_df.loc[indices[0], 'longitude'], temp_df.loc[indices[0], 'latitude'])
plt.scatter(user[:, 0], user[:, 1])
plt.legend(["member", "recommendation", "user"])
```

## Recommendation with Nearest Neighbors

```python
def recommend_byloc(user_id, n_recommendations=5):
    temp = run_df.loc[run_df.userId != user_id]
    
    X = temp.loc[:, ['longitude', 'latitude']] # scaled_run_df.loc[scaled_run_df.userId != user_id, ['longitude', 'latitude']]

    model = NearestNeighbors(n_neighbors=n_recommendations).fit(X)
#     model.fit(X.loc[:, ['longitude', 'latitude']])
    
    plt.scatter(X.longitude, X.latitude)
    plt.show()
    
    user = run_df.loc[run_df.userId == user_id, ['longitude', 'latitude']].mean()
    print(user)
    
    predictions = model.kneighbors(user.values.reshape(1, -1))
    return predictions, X, user
```

```python
(distances, indices), res_df, user = recommend_byloc(5850177, 10)
distances, indices
```
